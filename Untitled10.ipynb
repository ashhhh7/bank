```python
import pandas as pd

Data = pd.read_csv('/content/diabetes - diabetes.csv')

# Basic Statistics
print(Data.describe())

# Mean, Median, Mode
print("Mean:\n", Data.mean())
print("Median:\n", Data.median())
print("Mode:\n", Data.mode().iloc[0])

# Variance and Standard Deviation
print("Variance:\n", Data.var())
print("Standard Deviation:\n", Data.std())

# Skewness and Kurtosis
print("Skewness:\n", Data.skew())
print("Kurtosis:\n", Data.kurtosis())
```
